### Import the necessary libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Load each country’s cleaned dataset

```python
# Load cleaned datasets correctly
benin = pd.read_csv(r"C:\Users\Dell\Pictures\Solar-challenge-week0\Solar-challenge-week0\data\benin-clean.csv")
sierra_leone = pd.read_csv(r"C:\Users\Dell\Pictures\Solar-challenge-week0\Solar-challenge-week0\data\sierraleone-clean.csv")
togo = pd.read_csv(r"C:\Users\Dell\Pictures\Solar-challenge-week0\Solar-challenge-week0\data\togo-clean.csv")
```

### Display the first few rows of each dataset

```python

print("Benin Data:")
display(benin.head())

print("\nSierra Leone Data:")
display(sierra_leone.head())

print("\nTogo Data:")
display(togo.head())
```

### Add a “Country” column to each dataset

```python
benin["Country"] = "Benin"
sierra_leone["Country"] = "Sierra Leone"
togo["Country"] = "Togo"
```

### Combine all three into one DataFrame

```python
combined = pd.concat([benin, sierra_leone, togo], ignore_index=True)
```

### Import plotting libraries

```python
sns.set(style="whitegrid")
```

### Create boxplots for each metric

```python
# Boxplot for GHI
plt.figure(figsize=(8, 5))
sns.boxplot(x="Country", y="GHI", data=combined, palette="Set2")
plt.title("Global Horizontal Irradiance (GHI) Comparison")
plt.xlabel("Country")
plt.ylabel("GHI (W/m²)")
plt.show()

# Boxplot for DNI
plt.figure(figsize=(8, 5))
sns.boxplot(x="Country", y="DNI", data=combined, palette="Set2")
plt.title("Direct Normal Irradiance (DNI) Comparison")
plt.xlabel("Country")
plt.ylabel("DNI (W/m²)")
plt.show()

# Boxplot for DHI
plt.figure(figsize=(8, 5))
sns.boxplot(x="Country", y="DHI", data=combined, palette="Set2")
plt.title("Diffuse Horizontal Irradiance (DHI) Comparison")
plt.xlabel("Country")
plt.ylabel("DHI (W/m²)")
plt.show()
```

### Select metrics and country

```python
# Metrics to summarize
metrics = ["GHI", "DNI", "DHI"]

# Group by country and calculate mean, median, and std
summary = combined.groupby("Country")[metrics].agg(["mean", "median", "std"])

# Optional: round for readability
summary = summary.round(2)

# Display
print(summary)
```

```python
from tabulate import tabulate

print(tabulate(summary, headers='keys', tablefmt='github'))


```

```python
from scipy import stats
```

### Prepare GHI values by country

```python
ghi_benin = combined[combined["Country"] == "Benin"]["GHI"]
ghi_sierra = combined[combined["Country"] == "Sierra Leone"]["GHI"]
ghi_togo = combined[combined["Country"] == "Togo"]["GHI"]
```

### Run One-Way ANOVA

```python
f_stat, p_value = stats.f_oneway(ghi_benin, ghi_sierra, ghi_togo)
print("One-way ANOVA:")
print(f"F-statistic: {f_stat:.3f}, p-value: {p_value:.4f}")
```

### (Optional) Run Kruskal–Wallis test if ANOVA assumptions may not hold

```python
h_stat, p_value_kw = stats.kruskal(ghi_benin, ghi_sierra, ghi_togo)
print("\nKruskal–Wallis test:")
print(f"H-statistic: {h_stat:.3f}, p-value: {p_value_kw:.4f}")
```

### Cross-Country Solar Data Summary

* **Benin** records the highest average **GHI** (240.56 W/m²) and **DNI** (167.19 W/m²), along with slightly elevated **DHI** values. However, it also exhibits the greatest variability, reflecting periods of both very high and very low solar radiation.

* **Sierra Leone** shows the lowest mean **GHI** (201.96 W/m²) and **DNI** (116.38 W/m²), paired with relatively stable **DHI** levels. This suggests more consistent but generally lower solar energy potential.

* **Togo** lies between the two extremes, with moderate averages and variability across all three metrics, indicating a balanced solar profile with occasional peaks in **DHI**.



### Visual Summary using bar chart

```python

# Compute average GHI for each country
avg_ghi = {
    'Benin': benin['GHI'].mean(),
    'Sierra Leone': sierra_leone['GHI'].mean(),
    'Togo': togo['GHI'].mean()
}

# Create bar chart
plt.figure(figsize=(6,4))
plt.bar(avg_ghi.keys(), avg_ghi.values(), color=['tomato', 'skyblue', 'gold'])
plt.title('Average GHI by Country')
plt.ylabel('GHI (W/m²)')
plt.xlabel('Country')
plt.ylim(0, max(avg_ghi.values()) + 50)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
